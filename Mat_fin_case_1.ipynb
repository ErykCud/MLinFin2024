Usuwam outliery

```python
def remove_outliers_iqr(df):


  df_num = df.select_dtypes(include=[np.number])

  Q1 = df_num.quantile(0.20)
  Q3 = df_num.quantile(0.80)
  IQR = Q3 - Q1


  for index, row in df_num.iterrows():
    is_outlier = False
    for col in df_num.columns:
      if (row[col] > (Q3.loc[col] +  IQR.loc[col])) or (row[col] < (Q1.loc[col] -  IQR.loc[col])):
        is_outlier = True
        break

    if is_outlier:
      df.drop(index, inplace=True)

  return df

result_df = remove_outliers_iqr(df[nums].copy())
```

```python


def cap_outliers_iqr(df):
    df_num = df.select_dtypes(include=[np.number])
    Q1 = df_num.quantile(0.25)
    Q3 = df_num.quantile(0.75)
    IQR = Q3 - Q1

    for index, row in df_num.iterrows():
        for col in df_num.columns:
            upper_bound = Q3.loc[col] + 1.1 * IQR.loc[col]
            lower_bound = Q1.loc[col] - 0.9 * IQR.loc[col]
            df.loc[index, col] = np.clip(row[col], lower_bound, upper_bound)

    return df
result_df = cap_outliers_iqr(df[nums].copy())
```
